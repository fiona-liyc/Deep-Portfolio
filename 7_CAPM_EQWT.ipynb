## CAPM

```python
df.std()
```

```python
cov_matrix = asset_11.pct_change().cov()
```

```python
#from the calculations in the CAPM Excel
weights = [0.3690, 0.1167,0.0500, 0.0943, 0.0360, 
          0.0741, 0.0318, 0.0287, 0.0414, 0.1376, 0.0204]
```

```python
np.sum(weights)
```

```python
std = np.sqrt(np.dot(weights, np.dot(cov_matrix, weights))) * np.sqrt(52)
```

```python
print(std)
```

```python
rf=0.01
exp_ret = 0.06008
```

```python
sharpe_ratio = (exp_ret-rf)/std
print(sharpe_ratio)
```

## EQWT

```python
weekly_returns = df.pct_change().dropna()
```

```python
weekly_returns.head()
```

```python
weights2 = [0.0909,0.0909,0.0909,0.0909,0.0909,0.0909,0.0909,0.0909,0.0909,0.0909,0.0909]
```

```python
exp_ret = np.sum((weekly_returns.mean()*weights2)*52)
```

```python
exp_ret
```

```python
exp_std = np.sqrt(np.dot(weights2,np.dot(weekly_returns.cov()*52,weights2)))
```

```python
print(exp_std)
```

```python
sharpe_ratio = (exp_ret-rf)/exp_std
print(sharpe_ratio)
```
