```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#stock data input: only needs date & adjusted close price
data_vix = pd.read_csv('Data/VIX.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_ivv = pd.read_csv('Data/IVV.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_lqd = pd.read_csv('Data/LQD.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_shy = pd.read_csv('Data/SHY.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_tip = pd.read_csv('Data/TIP.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_vig = pd.read_csv('Data/VIG.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_vnq = pd.read_csv('Data/VNQ.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_gld = pd.read_csv('Data/GLD.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_efa = pd.read_csv('Data/EFA.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_eem = pd.read_csv('Data/EEM.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_emb = pd.read_csv('Data/EMB.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_mub = pd.read_csv('Data/MUB.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')

data_xle = pd.read_csv('Data/XLE.csv', header=0,usecols=['Date','Adj Close'], 
                   parse_dates=True, index_col='Date')
```

```python
market_econ_data = pd.concat([data_vix,data_ivv,data_lqd,data_shy,data_tip,data_vig,data_vnq,data_gld,data_efa,data_eem,data_emb,data_mub,data_xle],axis=1)
```

```python
market_econ_data.head()
```

```python
market_econ_data.columns = ['VIX_adj_close','IVV_adj_close','LQD_adj_close','SHY_adj_close','TIP_adj_close','VIG_adj_close','VNQ_adj_close','GLD_adj_close','EFA_adj_close','EEM_adj_close','EMB_adj_close','MUB_adj_close','XLE_adj_close']
```

```python
market_econ_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#transform the data to matrix form - to ensure the separation of scalar data
#[0,1]
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(market_econ_data.as_matrix())
print(scaled_data)
```

```python
#13 obs, 13 columns
np.shape(scaled_data)
```

```python
plt.plot(scaled_data)
plt.legend()
```

```python
##PCA##
from sklearn.decomposition import PCA
```

```python
# pending: research on N_compo to find optimal components
    #storing data in 4 columns
    #these 4 lines almost explain all the data
    #reduce storage, easy retrieval of original data
    #the more component the better, there always an optimal value
pca = PCA(n_components=4)
pca_component = pca.fit_transform(scaled_data)
```
